```python
full_year = []

##loop to check adherence column to determine if early, late, or on time append

for x in pop_income['YEAR']:
    if x ==3:
        full_year.append('2010')
    elif x == 4:
        full_year.append('2011')
    elif x == 5:
        full_year.append('2012')
    elif x == 6:
        full_year.append('2013')
    elif x == 7:
        full_year.append('2014')
    elif x == 8:
        full_year.append('2015')
    elif x == 9:
        full_year.append('2016')
    elif x == 10:
        full_year.append('2017')
    elif x == 11:
        full_year.append('2018')
    elif x == 12:
        full_year.append('2019')
    else:
        full_year.append('NaN')
        
pop_income['full_year']  = full_year
```
